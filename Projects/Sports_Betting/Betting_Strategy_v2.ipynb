```python
def beatTheBookie(data, bet, marg):
    
    money_list = [] ##initial amount of money
    money = 0
    i = 0 # counter for money
    accuracy = []
    acc = 1
    nGames = data.shape[0]
    bet_games = []
    nValidOdds = 3
    # bet_result = []
    max_odds = []
    ids = []
    mean_odds = []


    for gm in range(nGames):

        ## Comparing the scores for home vs away
        score_home = data.loc[gm, "score_home"]
        score_away = data.loc[gm, "score_away"]

        if score_home > score_away:
            result = 1

        elif score_home == score_away:
            result = 2

        else:
            result = 3

        ##averages  

        averages = []

        filter_col_home = [col for col in data if col.startswith('home')]
        home_df = data.loc[gm,filter_col_home].copy()
        avg_home = home_df.mean()
        averages.append(avg_home)

        filter_col_draw = [col for col in data if col.startswith('draw')]
        draw_df = data.loc[gm,filter_col_draw].copy()
        avg_draw = draw_df.mean()
        averages.append(avg_draw)


        filter_col_away = [col for col in data if col.startswith('away')]
        away_df = data.loc[gm,filter_col_away].copy()
        avg_away = away_df.mean()
        averages.append(avg_away)

        ##maximums [max_home max_draw max_away];

        max_home = home_df.max()
        max_draw = draw_df.max()
        max_away = away_df.max()

        maximums = [max_home, max_draw, max_away]     


        count_home = home_df.count()
        count_draw = draw_df.count()
        count_away = away_df.count()

        counts = [count_home, count_draw, count_away]

        ## check that there is at least one valid bet. If the bets are still not
        ## online for this time, continue to the next game

        if (any(pd.isna(averages)) and  any(pd.isna(maximums))):
            continue


        ##Apply formula and decide whether to bet
        earn_margin = [0,0,0]

        earn_margin[0] = ((1 / averages[0] - marg) * maximums[0] - 1) * (counts[0] > nValidOdds) 
        earn_margin[1] = ((1 / averages[1] - marg) * maximums[1] - 1) * (counts[1] > nValidOdds) 
        earn_margin[2] = ((1 / averages[2] - marg) * maximums[2] - 1) * (counts[2] > nValidOdds) 

        if (sum(x for x in earn_margin if x > 0) >= 1):

            bet_result = earn_margin.index(max(earn_margin)) + 1

            possible_earn = bet  * (maximums[bet_result-1] - 1) ## look into it

            max_odds.append(maximums[bet_result-1]) #check
            mean_odds.append(averages[bet_result-1]) #ok
            #ids.append(gm)

            ## calculate loss / earning
            if (bet_result ==  result):
                money = money + possible_earn
                money_list.append(money)
                accuracy.append(1)
            else:
                money = money - bet
                money_list.append(money)
                accuracy.append(0)

            bet_games.append(gm)
    
# Ask

    ##### returns 
    # 1. total money, 
    # 2. list of returns from all placed bets
    # 3. id of games on which bets were placed
    # 4. list of max odds
    # 5. list of mean odds
    # 6. boolean list showing which bets worked and which didn't

    return (money,money_list,bet_games,max_odds,mean_odds,accuracy)
 
```
